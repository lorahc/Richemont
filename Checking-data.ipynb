```python
import os
import wave
import pandas as pd
```

```python
#number of folders

base_path = "/Users/laura/Documents/RICHEMONT/250319/"

# Carpetas a revisar dentro de cada nombre
subfolders = ["Emotibit", "Audio", "Polar ECG"]

def bytes_to_human_readable(num_bytes):
    # Función para convertir bytes a KB, MB, GB para que sea más legible
    for unit in ['B','KB','MB','GB','TB']:
        if num_bytes < 1024:
            return f"{num_bytes:.2f} {unit}"
        num_bytes /= 1024
    return f"{num_bytes:.2f} PB"

for person_folder in os.listdir(base_path):
    person_path = os.path.join(base_path, person_folder)
    if os.path.isdir(person_path):
        print(f"\FOLDER: {person_folder}")
        for subfolder in subfolders:
            folder_path = os.path.join(person_path, subfolder)
            if os.path.exists(folder_path) and os.path.isdir(folder_path):
                files = os.listdir(folder_path)
                num_files = len(files)
                total_size = 0
                for f in files:
                    file_path = os.path.join(folder_path, f)
                    if os.path.isfile(file_path):
                        total_size += os.path.getsize(file_path)
                print(f"  {subfolder}: {num_files} files, {bytes_to_human_readable(total_size)}")
       
```

```python


base_path = "/Users/laura/Documents/RICHEMONT/250319/"
sample_rate = 16000  #
for person_folder in os.listdir(base_path):
    person_path = os.path.join(base_path, person_folder)
    if os.path.isdir(person_path):
        print(f"\n=== {person_folder} ===")

        # Audio
        audio_path = os.path.join(person_path, "Audio")
        if os.path.exists(audio_path):
            print(" Audio files:")
            for f in os.listdir(audio_path):
                file_path = os.path.join(audio_path, f)
                if os.path.isfile(file_path) and f.lower().endswith(".wav"):
                    try:
                        with wave.open(file_path, 'rb') as wav_file:
                            frames = wav_file.getnframes()
                            duration_min = frames / sample_rate / 60
                            print(f"  {f}: {duration_min:.2f} minutes")
                    except Exception as e:
                        print(f"  [ERROR WAV] {f} - {str(e)}")
        else:
            print("  No Audio folder found.")

        #Polar ECG
        polar_path = os.path.join(person_path, "Polar ECG")
        if os.path.exists(polar_path):
            print(" Polar ECG files:")
            for f in os.listdir(polar_path):
                file_path = os.path.join(polar_path, f)
                if os.path.isfile(file_path) and f.lower().endswith(".csv"):
                    try:
                        df = pd.read_csv(file_path)
                        df['real_time'] = pd.to_datetime(df.iloc[:, 0], unit='ns')
                        start_time = df['real_time'].min()
                        end_time = df['real_time'].max()
                        duration_min = (end_time - start_time).total_seconds() / 60
                        print(f"  {f}:")
                        print(f"    Start time: {start_time}")
                        print(f"    End time:   {end_time}")
                        print(f"    Duration:   {duration_min:.2f} minutes")
                    except Exception as e:
                        print(f"  [ERROR CSV] {f} - {str(e)}")
        else:
            print("  No Polar ECG folder found.")

```

```python
# checkin emotibit duration
import os
import pandas as pd
from datetime import datetime, timedelta


time_sync_paths = [
    "/Users/laura/Documents/RICHEMONT/250319/Clement/Emotibit/Parsed data 2/2025-03-19_10-19-19-735210_timeSyncMap.csv",
  
    
]
metric_to_check = "PG"

for sync_path in time_sync_paths:
    if not os.path.exists(sync_path):
        print(f"Missing folder: {sync_path}")
        continue

    folder = os.path.dirname(sync_path)
    person = sync_path.split(os.sep)[-4]
    parsed_folder = sync_path.split(os.sep)[-2]

    print(f"\n=== {person} ===")
    print(f"  Folder: {parsed_folder}")

    df_map = pd.read_csv(sync_path)
    TE0 = float(df_map['TE0'].iloc[0])
    TL0 = float(df_map['TL0'].iloc[0])


    start_time_str = os.path.basename(sync_path).replace("_timeSyncMap.csv", "")
   
    start_time_host = datetime.strptime(start_time_str, "%Y-%m-%d_%H-%M-%S-%f")
   


    anchor_time = start_time_host + timedelta(seconds=TL0 / 1e9)

    print(f"    TE0: {TE0}")
    print(f"    TL0 (nanosecs): {TL0}")
    print(f"    Anchor time: {anchor_time}")


    pg_file = next((f for f in os.listdir(folder) if f.endswith(f"_{metric_to_check}.csv")), None)
    if not pg_file:
        print(f"    No file {metric_to_check}.")
        continue

    df = pd.read_csv(os.path.join(folder, pg_file))

    print(f"    Primer timestamp EmotiBitTimestamp: {df['EmotiBitTimestamp'].iloc[0]}")
    print(f"    Último timestamp EmotiBitTimestamp: {df['EmotiBitTimestamp'].iloc[-1]}")

    if df['EmotiBitTimestamp'].max() > 1e6:
        
        df["real_time"] = df["EmotiBitTimestamp"].apply(
            lambda x: anchor_time + timedelta(seconds=(x / 1000))
        )
    else:
    
        df["real_time"] = df["EmotiBitTimestamp"].apply(
            lambda x: anchor_time + timedelta(seconds=x)
        )

    start = df["real_time"].iloc[0]
    end = df["real_time"].iloc[-1]

    print(f"    Start time real: {start}")
    print(f"    End time real:   {end}")

    
```
